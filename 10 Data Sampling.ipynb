##### Stratified Sampling

```python
import pandas as pd
```

```python
df = pd.read_csv('loan-prediction.csv')
```

```python
df.shape
```

```python
x = df.drop(['Loan_ID','Loan_Status'], axis = 1)
y = df['Loan_Status']
```

```python
y
```

```python
plt.pie(y.value_counts(), labels=y.value_counts().index, autopct='%2.2f%%');
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x.shape
```

```python
x_train.shape, x_test.shape
```

```python
plt.subplot(1,3,1)
plt.title('Original')
plt.pie(y.value_counts(), labels=y.value_counts().index, autopct='%2.2f%%');
plt.subplot(1,3,2)
plt.title('Training')
plt.pie(y_train.value_counts(), 
        labels=y_train.value_counts().index, autopct='%2.2f%%');
plt.subplot(1,3,3)
plt.title('Testing')
plt.pie(y_test.value_counts(), 
        labels=y_test.value_counts().index, autopct='%2.2f%%');
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, random_state=0, stratify=y)
```

```python
plt.subplot(1,3,1)
plt.title('Original')
plt.pie(y.value_counts(), labels=y.value_counts().index, autopct='%2.2f%%');
plt.subplot(1,3,2)
plt.title('Training')
plt.pie(y_train.value_counts(), 
        labels=y_train.value_counts().index, autopct='%2.2f%%');
plt.subplot(1,3,3)
plt.title('Testing')
plt.pie(y_test.value_counts(), 
        labels=y_test.value_counts().index, autopct='%2.2f%%');
```

##### Random Under Sampler

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=0)
```

```python
x_new, y_new = rus.fit_resample(x, y)
```

```python
y_new.shape
```

```python
plt.pie(y_new.value_counts(), labels=y_new.value_counts().index, autopct='%2.2f%%');
```

##### Random Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
```

```python
x_new, y_new = ros.fit_resample(x, y)
```

```python
y_new.shape
```

```python
plt.pie(y_new.value_counts(), labels=y_new.value_counts().index, autopct='%2.2f%%');
```

##### SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smo = SMOTE(random_state=0)
```

```python
x_new, y_new = smo.fit_resample(x, y)
```

```python
plt.pie(y_new.value_counts(), labels=y_new.value_counts().index, autopct='%2.2f%%');
```
